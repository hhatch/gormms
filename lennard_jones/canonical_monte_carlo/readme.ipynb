# Canonical Ensemble Metropolis Monte Carlo Simulations



```python
import pandas as pd
from tabulate import tabulate
from dormms import lennard_jones as lj
data = lj.CanonicalMonteCarlo().data()
#print(data)
print('Simulation parameters')
tabulate([[r'$N$ :', data['number_particles']],
          [r'$r_c$ :', str(data['interaction_cutoff']) + ' + long range corrections'],
          ['Equilibration trials :', data['equilibration_trials']],
          ['Production trials :', "{:e}".format(data['production_trials'])],
               ], tablefmt='html')
```

```python
pd.DataFrame({r'$k_B T/\epsilon$': data['temperature'],
              r'$\rho\sigma^3$': data['density'],
              r'$U/\epsilon$': data['energy_total'],
              r'$\sigma_{U}/\epsilon$': data['energy_total_blockstdev']})
```
